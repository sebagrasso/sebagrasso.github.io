####  2.1 New Operations

##### 2.1.1 Header

```python
# OPERATIONS UPDATE?????

flag_buy = False
flag_sell = False

# flag_buy = True
# flag_sell = False

tick = 'AAPL'
date_oper = "2022-09-20"
# date_oper = datetime.now().strftime("%Y-%m-%d")

if flag_buy:
    flag_operations = True
    ###### Amount invested in eur. BUY: - , SELL = +
    try:
        # TICK COMPLETLY SOLD SOMETIME. THEN BUY THAT WHOLE AMOUNT
        net_eur = -operations_df[(operations_df.ticker == tick) & (operations_df["net_(eur)"] > 0 )]["net_(eur)"][-1]
    except:
        #TICK NEVER SOLD. 
        net_eur = -100

elif flag_sell:
    flag_operations = True
    ###### Amount invested in shares. BUY: + , SELL = -
    # SELLING ALL SHARES
    shares = -operations_df.groupby("ticker").sum()['shares'][tick]
else:
    flag_operations = False
```

##### 2.1.2 Available currencies conversion DF

```python
oper_curr = operations_df["currency"].str.upper().unique()
oper_curr = oper_curr[~np.isin(oper_curr,"EUR")]
eur_to_conv_df = pd.DataFrame()

for oc in oper_curr:
    curr = "EUR"+oc+"=X"
    conv = yf.Ticker(curr).history(period='10y', rounding=True).iloc[:,3]
    conv.name = oc
    conv = pd.concat([pd.DataFrame(pd.date_range(start = conv.index.min(), end = today)).set_index(0),conv],
                       axis=1, join="outer").pad()
    conv.index.name = "Date"
    conv.columns.name = curr
    eur_to_conv_df = pd.concat([eur_to_conv_df,conv],axis=1)
    
eur_to_conv_df["EUR"] = 1
```

##### 2.1.3 New Row to append to Operations DF

```python
if flag_operations:
    currency = si.get_quote_data(tick)['currency']
    operation_price = data.resample("1d").mean()[data.resample("1d").mean().index == date_oper].loc[:,(tick,"Close")][0]
    operation_price_eur = operation_price / eur_to_conv_df.loc[date_oper,currency.upper()]
    
    current_price = data.resample("1d").mean().loc[:,(tick,"Close")][-1]
    current_price_eur = current_price / eur_to_conv_df.loc[today,currency.upper()]
    new_oper_net = [-operation_price_eur*shares if flag_sell else net_eur][0]
    
    try:
        # TWO SELLS OF THE TICKER
        dates_between_sells = (operations_df.index > operations_df[(operations_df.ticker == tick) & (operations_df["sales_num"] == 1)].index[-2]) & \
                              (operations_df.index < operations_df[(operations_df.ticker == tick) & (operations_df["sales_num"] == 1)].index[-1])
    except:
        try:
            # ONE SELLS OF THE TICKER
            dates_between_sells = (operations_df.index < operations_df[(operations_df.ticker == tick) & (operations_df["sales_num"] == 1)].index)
        except:
            # TICKER NEVER SOLD
            dates_between_sells = operations_df[(operations_df.ticker == tick)].index
    net_bought = operations_df[(np.isin(operations_df.index,dates_between_sells)) & (operations_df.ticker == tick)]["net_(eur)"].sum()
    
    new_oper = {"Date":date_oper,
                "ticker":tick,
                "net_(eur)":new_oper_net,
                "name":si.get_quote_data(tick)['shortName'],
                "exchange":si.get_quote_data(tick)["fullExchangeName"],
                "country":si.get_quote_data(tick)['market'].split("_")[0],
                "currency":currency,
                "sales_num":[1 if new_oper_net>0 else 0][0],
                "purchases_num":[1 if new_oper_net<0 else 0][0],
                "operation_price": operation_price,
                "operation_price_(eur)": operation_price_eur,
                "shares": [shares if flag_sell else -new_oper_net / operation_price_eur][0],
                "returns_(eur)": [new_oper_net + net_bought if flag_sell else -new_oper_net / operation_price_eur * current_price_eur + new_oper_net],
                "current_price": current_price,
                "current_price_(eur)": current_price_eur
               }
    
    new_oper2 = pd.DataFrame(new_oper, index=[0]).set_index("Date", drop = True)
    new_oper2.index = pd.DatetimeIndex(new_oper2.index)
    operations_df = pd.concat([operations_df,new_oper2],axis=0)
    
    print(f"{tick} added to operations_df")

###### DOWNLOAD NEW TICKS HISTORY AND MERGE TO DATA, IF IT IS NOT ALREADY GOTTEN
    if np.where(ticks == tick,1,0).sum() == 0:
        ticks = np.append(ticks,tick)
        ticks = np.unique(ticks)
    
        tick_df = yf.Ticker(tick).history(period=period, interval=interval, rounding=True)
        tick_df.index = tick_df.index.tz_convert("Europe/Lisbon")
        tick_df = tick_df.dropna(axis=0, how="all")
        tick_df = pd.concat([pd.DataFrame(pd.date_range(start = tick_df.index.min(),
                                                            end = datetime.now(pytz.timezone('Europe/Lisbon')))).set_index(0),
                                 tick_df],
                                axis=1, join="outer").pad()
        print("New tick history edited")
        
        ###### SET MULTIINDEX TO NEW TICK DF AND CONCAT WITH DATA PRICES DF
        ind = []
        for i in tick_df.columns:
            ind.append([tick,i])
        ind = pd.MultiIndex.from_tuples(tuple(ind))
        
        tick_df.columns = ind
        data = pd.concat([tick_df,data],axis=1)
        data.columns.names = ['Ticker','Stock Info']
        data.index.name = "Date"
        print("New tick history added to data")

    
    # FlOAT NUMBERS FORMAT. 1 DECIMAL
    operations_df[["net_(eur)", "operation_price","operation_price_(eur)","current_price" ,"current_price_(eur)"]] = operations_df[["net_(eur)", "operation_price","operation_price_(eur)","current_price" ,"current_price_(eur)"]].round(1)
    operations_df["returns_(eur)"] = operations_df["returns_(eur)"].round(2)
    
    # RETURNS FOR TICKS WITHOUT SHARES AND WHEN I BOUGHT THEM = 0
    no_shares = operations_df.groupby("ticker")["shares"].sum() == 0
    tic_no_shares = (no_shares).index[no_shares].values
    operations_df["returns_(eur)"][np.isin(operations_df["ticker"],tic_no_shares) & (operations_df["sales_num"] == 0)] = 0
    operations_df = operations_df.sort_values(by="Date")
else:
    print("There is no new activity")
```

#### 2.2 Operation Prices (Done)

```python
if inter2 == 3:
    aux = pd.DataFrame(data.loc[:,(operations_df.ticker,"Close")].unstack()).reset_index().set_index("Date").drop(labels = "Stock Info", axis=1).rename(columns={"Ticker":"ticker", 0:"operation_price"})
    aux.index = pd.DatetimeIndex(aux.index.strftime("%Y-%m-%d"))
    
    # REFRESH PRICE OF ALL OPERATIONS
    operations_df["operation_price"] = pd.merge(operations_df[["ticker","operation_price"]], aux, how = "left", on = ["Date","ticker"]).iloc[:,-1]
    operations_df["operation_price_(eur)"] = np.round((operations_df["operation_price"].values / np.diag(eur_to_conv_df.loc[operations_df.index,operations_df["currency"].str.upper()])),2)
    operations_df["shares"] = -operations_df["net_(eur)"] / operations_df["operation_price_(eur)"]
else:
    print("Operations prices cannot be updated because history interval is not a day")
```

####  2.3 Operations Prices Update

```python
# UPDATE CURRENT PRICE (IN LOCAL CURRENCY)
operations_df["current_price"] = pd.merge(operations_df[["ticker","current_price"]],
             data.resample("1d").mean()[data.resample("1d").mean().index == today].loc[:,(ticks,"Close")].T.reset_index().rename(columns={"Ticker":"ticker"}),
             how="left",on="ticker").iloc[:,-1].values.round(1)

# UPDATE CURRENT PRICE (IN EUR)
operations_df["current_price_(eur)"] = np.round((operations_df["current_price"].values / eur_to_conv_df.loc[today,operations_df["currency"].str.upper()].values),1)

print("Current prices Updated")
```

####  2.4 Returns Update

```python
never_sold = operations_df.groupby("ticker")["sales_num"].sum() == 0
tic_never_sold = (never_sold).index[never_sold].values

# RETURNS FOR NEVER SOLD TICKS
mask_neversold = np.isin(operations_df["ticker"],tic_never_sold)

operations_df["returns_(eur)"][mask_neversold] = \
np.round(operations_df["shares"][mask_neversold] * \
         operations_df["current_price_(eur)"][mask_neversold] + \
         operations_df["net_(eur)"][mask_neversold],2)

#RETURNS FOR TICKS WITH SHARES AND WHEN I LAST BOUGHT THEM (SOLD AT LEAST ONCE)
try:
    mask_sishares_sold = ~np.isin(operations_df["ticker"],tic_no_shares) & (operations_df["sales_num"] == 1)
    #TICKER FROM NEW OPERATION, FOR DATES AFTER LAST SOLD
    dates_post_sell_sishares_sold = (operations_df.index > mask_sishares_sold.index[mask_sishares_sold][-1]) & (operations_df["sales_num"] == 0)
    
    operations_df["returns_(eur)"][dates_post_sell_sishares_sold] = \
    np.round(operations_df["shares"][dates_post_sell_sishares_sold] * \
             operations_df["current_price_(eur)"][dates_post_sell_sishares_sold] + \
             operations_df["net_(eur)"][dates_post_sell_sishares_sold],2)
except:
    pass
```
