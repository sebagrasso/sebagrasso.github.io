```python
def mat_corr(matcorr,target,coef):
    ### CORRELATION MATRIX FILTERED BY COEF
    matcorr_coef = matcorr[target][(matcorr[target] > coef) | (matcorr[target] < -coef)].dropna(how="all")
    ticks_corr = matcorr_coef.index.get_level_values(0).unique()
    ticks_corr = np.delete(ticks_corr,np.where(ticks_corr == target))
    ### drop target from matrix
    matcorr_coef = matcorr_coef.loc[(ticks_corr)]
    
    ### Create corr df
    return pd.DataFrame(matcorr_coef.max(1), columns = ["Corr"]), ticks_corr

def chart_pearson(target, Ticker, vs, stock_info, coef, i):

    if (data2y_ret.columns.get_level_values(0) == Ticker.upper()).sum() != 0:
        target = Ticker.upper()
        
    df_corr_coef, ticks_corr = mat_corr(matcorr_data2y_ret,target,coef)
    
    if (data2y_ret.columns.get_level_values(0) == vs.upper()).sum() != 0:
        vs = vs.upper()
    else:
        vs = df_corr_coef.index[i][0]
    
    if stock_info:
        pass
    else:
        stock_info = df_corr_coef.index[i][1]
        
    corr = df_corr_coef.loc[(vs,stock_info)][0].round(2)
    x = data2y_ret[target][stock_info]
    y = data2y_ret[vs][stock_info]
    
    plt.figure(figsize = (24,12))
    
    ax1 = plt.subplot(3,5,(1,8))
    ax2 = plt.subplot(3,5,(4,5))
    ax3 = plt.subplot(3,5,(9,10))
    ax4 = plt.subplot(3,5,(11))
    ax5 = plt.subplot(3,5,(13))
    ax6 = plt.subplot(3,5,(15))
    
    sns.regplot(x,y, ax=ax1)
    ax1.set_title(f"Returns correlation: {stock_info} | {target} vs {vs}", fontsize=35,pad=20)
    ax1.set_xlabel(target)
    ax1.set_ylabel(vs)
    ax1.set_xlim(np.min(x),np.max(x))
    ax1.set_ylim(np.min(x),np.max(x))
    ax1.annotate(f"Number of features: {df_corr_coef.size}",xy=(np.min(x)*.95,np.max(x)*.85), fontsize=25)    
    ax1.annotate(f"{data_gral[data_gral.ticker == vs].name.values[0]} | {data_gral[data_gral.ticker == vs].industry.values[0]} | {data_gral[data_gral.ticker == vs].country.values[0]}",
                 xy=(np.min(x)*.95,np.max(x)*.7), fontsize=25)
    ax1.annotate(f"Corr Per: {corr}",xy=(np.min(x)*.95,np.max(x)*.55), fontsize=20)
    
    ax2.plot(data2y[target][stock_info],"r")
    ax22 = ax2.twinx()
    ax22.plot(data2y[vs][stock_info],"y", label = vs)
    ax22.grid(False)
    ax2.set_title(f"{stock_info} Price | {target}-{vs}", fontsize = 20, pad = 10)
    
    ax3.plot(data2y_ret[target][stock_info],"g")
    ax3.plot(data2y_ret[vs][stock_info],"y", alpha= 0.7, label = vs)
    ax3.set_title(f"{stock_info} returns | {target}-{vs}", fontsize = 20, pad = -10)

    ax4.pie(data_gral[data_gral["ticker"].isin(ticks_corr)].loc[:,"sector"].value_counts(),
            labels=data_gral[data_gral["ticker"].isin(ticks_corr)].loc[:,"sector"].value_counts().index,
            wedgeprops={'linewidth': 2, 'edgecolor': 'white'},
            autopct=lambda p: format(p,".0f") if p > 10 else None)
    ax4.set_title("Sector", fontsize=18, pad = -10)
    
    ax5.pie(data_gral[data_gral["ticker"].isin(ticks_corr)].loc[:,"industry"].value_counts(),
            labels=data_gral[data_gral["ticker"].isin(ticks_corr)].loc[:,"industry"].value_counts().index,
            wedgeprops={'linewidth': 2, 'edgecolor': 'white'},
            autopct=lambda p: format(p,".0f") if p > 10 else None)
    ax5.set_title("Industry", fontsize=18, pad = -10)

    ax6.pie(data_gral[data_gral["ticker"].isin(ticks_corr)].loc[:,"country"].value_counts(),
            labels=data_gral[data_gral["ticker"].isin(ticks_corr)].loc[:,"country"].value_counts().index,
            wedgeprops={'linewidth': 2, 'edgecolor': 'white'},
            autopct=lambda p: format(p,".0f") if p > 10 else None)
    ax6.set_title("Country", fontsize=18, pad = -10)
    # plt.tight_layout()
    plt.show()
    
target = widgets.Dropdown(options = ticks_strategy, layout=widgets.Layout(width="10%"))

stock_info = widgets.Dropdown(options = ohlcv_cols, layout=widgets.Layout(width="10%"))

coef = widgets.FloatSlider(min = 0, max = 1, step = 0.1, value = 0.6,
                           description = "Corr cut-off", layout=widgets.Layout(width="30%"))
i = widgets.IntSlider(min=0, max=5727, description="Feature nº", layout=widgets.Layout(width="30%"))

Ticker = widgets.Text(value = "",
                       style = {'description_width': 'initial'}, layout=widgets.Layout(width="20%"))

vs = widgets.Text(value = "",
                       style = {'description_width': 'initial'}, layout=widgets.Layout(width="20%"))

ui = widgets.Box([target, Ticker, vs, stock_info, coef, i])
out = widgets.interactive_output(chart_pearson, {'target': target, "Ticker":Ticker, "vs":vs, "stock_info":stock_info, 'coef': coef, 'i':i})
display(ui,out)
```

```python
def sigma_analy(tic, Ticker, stock_info, start, func):
    
    if (data2y_ret.columns.get_level_values(0) == Ticker.upper()).sum() != 0:
        tic = Ticker.upper()
  
    df = data2y
    df_ret = data2y_ret
    df_ret_out, filt, rango_min, rango_max = func(df_ret)
    
    plt.figure(figsize=(20,8))
    ax1 = plt.subplot(6,3,(1,5))
    ax2 = plt.subplot(6,3,(7,11))
    ax3 = plt.subplot(6,3,(13,17))
    ax4 = plt.subplot(6,3,(3,9))
    ax5 = plt.subplot(6,3,(12,18))
    
    x = df_ret.loc[:,(tic,stock_info)][start:]
    x2 = df.loc[:,(tic,stock_info)][start:]
    y = df_ret_out.loc[:,(tic,stock_info)][start:]
    z = filt.loc[:,(tic,stock_info)][start:]
    
    ax1.set_title(f"{tic} - {stock_info}", fontsize = 30)
    ax1.plot(x2 , label = "Stock_info")
    ax1.get_xaxis().set_visible(False)
    ax1.legend()
    
    ax2.plot(x, label = "Returns")
    ax2.set_ylim(np.min(x),np.max(x))
    ax2.scatter(z.index, z, color="red", label = "Outliers")
    ax2.get_xaxis().set_visible(False)
    ax2.legend()
    
    # Tic Index to plot rango_min and rango_max    
    ind = np.where(df_ret.columns == (tic,stock_info))
    ax2.axhline(rango_min[ind][0], color = "orange", linestyle = "dashed")
    ax2.axhline(rango_max[ind][0], color = "brown", linestyle = "dashed")
        
    ax3.plot(y,c="r", label = "Returns without Outliers")
    ax3.set_ylim(np.min(y),np.max(y))
    ax3.legend()
    
    sns.distplot(x, bins=50, label="Returns",ax=ax4)
    sns.distplot(y, bins=50, label="Returns without Outliers",ax=ax4)
    ax4.set_xlim(np.percentile(x,0.5),np.percentile(x,99.5))
    ax4.set_ylim(np.percentile(x,75))
    ax4.legend()
    
    ax5.boxplot([x,y],labels=["Original","Transformed"])
    plt.tight_layout()
    plt.show()

start = widgets.IntSlider(min = 0 , max = data2y_ret.index.size-1, step = 60, value = 700, description = "Steps before",
                                  style = {'description_width': 'initial'},
                                 layout=widgets.Layout(width="40%"))

stock_info = widgets.Dropdown(options = ohlcv_cols, value = "Close",
                       style = {'description_width': 'initial'}, layout=widgets.Layout(width="10%"))

tic = widgets.Dropdown(options = std_data2y_ret.index.get_level_values(0),
                      style = {'description_width': 'initial'}, layout=widgets.Layout(width="10%"))

func = widgets.Dropdown(options = [outliers_sigma,outliers_iqr],
                       style = {'description_width': 'initial'}, layout=widgets.Layout(width="40%"))

Ticker = widgets.Text(value = "",
                       style = {'description_width': 'initial'}, layout=widgets.Layout(width="40%"))

ui = widgets.Box([tic,Ticker,stock_info,start,func])
out = widgets.interactive_output(sigma_analy, {'tic': tic, "Ticker":Ticker, 'stock_info': stock_info, 'start': start, "func":func})
display(ui,out)
```
